###**Resumen**
A mediados del siglo XX, en la ciudad de Guadalajara se necesito realizar una ampliación de la Avenida Federalismo. Sin embargo, este proyecto tuvo un gran obstáculo, ya que el edificio de Telefonos de México de 1,700 toneladas y 5 pisos de altura, se interponía justo en la trayectoria planificada de la nueva avenida.

La solución convencional hubiera sido la demolición, pero esto era costoso, generaría mucho escombro y interrumpiría el servicio crítico de suministro eléctrico que el edificio proporcionaba. La solucuión que implemento en ese momento el ingeniero civil Jorge Matute Remus fue levanatar el edificio con gatos hidraulicos y moverlo 12 metros en 4 dias
